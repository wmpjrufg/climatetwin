# Dataset completo

```python
import os

from climate_twin import *

caminho_pasta = r'C:\git-projetos\climatetwin\BD\dados\wander'
excel = [arquivo for arquivo in os.listdir(caminho_pasta) if arquivo.endswith(('.csv', '.xlsx', '.xls'))]
arquivos_excel = [os.path.join(caminho_pasta, arquivo) for arquivo in excel]
chuvas = []
for i in arquivos_excel:
    metadados, dados = ler_dados(i)
    _, chuva = calculo_precipitacoes(dados, metadados)
    chuvas.append(chuva)
df_chuvas_brasil = pd.concat(chuvas, ignore_index=True)
df_chuvas_brasil
```

```python
df_chuvas_brasil.to_excel(r'C:\git-projetos\climatetwin\chuvas_brasil_completo.xlsx', index=False)
```
